````python
class IdentityTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, input_array, y=None):
        return self
    
    def transform(self, input_array, y=None):
        return input_array*1
````

````python
ToNormalStandarScaler = Pipeline([('yeo-johnson', PowerTransformer(method='yeo-johnson')),
                                  ('scaler', StandardScaler())
                                 ])

preprocessing_transformer = ColumnTransformer(
    transformers=[
        ('ReduceCardOneHotEncoder1', OneHotEncoder(sparse=False, min_frequency=0.03, handle_unknown='infrequent_if_exist'),  ['Nationality','Club_Position']),
        ('OneHotEncoder1', OneHotEncoder(sparse=False),  ['Work_Rate']),
        ('OneHotEncoder2', OneHotEncoder(sparse=False, drop='if_binary'),  ['Preffered_Foot']),
        ('Identity', IdentityTransformer(),  ['Weak_foot', 'Skill_Moves']),
        ('StandardScaler', StandardScaler(),  ['Height', 'Weight']),
        ('ToNormalStandarScaler', ToNormalStandarScaler, ['Age', 'Ball_Control', 'Dribbling',
                                                          'Marking', 'Sliding_Tackle', 'Standing_Tackle',
                                                          'Aggression', 'Reactions', 'Interceptions',
                                                          'Vision', 'Composure', 'Crossing',
                                                          'Short_Pass', 'Long_Pass', 'Acceleration',
                                                          'Speed', 'Stamina', 'Strength',
                                                          'Balance', 'Agility', 'Jumping',
                                                          'Heading', 'Shot_Power', 'Finishing',
                                                          'Long_Shots', 'Curve', 'Freekick_Accuracy',
                                                          'Penalties', 'Volleys'])
    ])


players_pipeline1 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', svm.SVC())
])

players_pipeline2 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', RandomForestClassifier())
])

players_pipeline3 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', LogisticRegression())
])

players_pipeline1_ros = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', svm.SVC())
])

players_pipeline2_ros = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', RandomForestClassifier())
])

players_pipeline3_ros = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', LogisticRegression())
])
````

Justificaci칩n transformaciones:

* El primer OHE se aplica a variables categ칩ricas con un n칰mero amplio de categor칤as, esto debido a que este OHE crea una categor칤a donde agrupa todas las categor칤as poco frecuentes en una. De esta manera se puede reducir en gran medida la dimensionalidad de los datos finales.

* El segundo OHE se aplica a otras variables categ칩ricas no binarias con un n칰mero reducido de opciones, a diferencia del anterior donde las opciones eran m치s de 20 en el caso de Posiciones. En este caso s칩lo se aplica a la columna work_rate.

* El 칰ltimo OHE se aplica a las variables categ칩ricas binarias, ya que al hacer la transformaci칩n a una variable binaria indicando el argumento "if_binary", se obtiene una sola columna con 0s para la primera clase y 1s para la otra (eliminando la necesidad de tener 2 columnas con esto mismo). 

* Luego, a modo de que todas las columnas necesarias est칠n disponibles, se agrega una transformaci칩n identidad. Esto para aquellos datos que no necesitan transformaci칩n como Weak_foot y Skill_moves.

* La pen칰ltima transformaci칩n es un standar scaler, que se aplica s칩lo a variables que se ve que distribuyen normal.

* Para aquellas variables que no distribuyen normal, se les aplica una transformaci칩n para adaptar las distribuciones seguido del standar scaler.

La precisi칩n indica cuantas predicciones de la clase positiva 1 (jugador es seleccionado nacional) son correctas. Para todos los casos sin over sampling es posible notar que la precisi칩n es muy malo o enga침ador. Para el caso de SVC su valor es de 0, pero no existe ninguna predicci칩n para la clase positiva. Por otro lado el Random Forest obtiene una precisi칩n de 1, pero es enga침ador porque observando la matriz de confusi칩n, apenas 2 predicciones fueron para la clase positiva, y calz칩 que ambas predicciones fueron correctas. Por 칰ltimo ocurre lo mismo para el Logistic Regressor, donde apenas 3 predicciones fueron de la clase positiva.

El Recall indica cuantos de los casos positivos fueron predichos correctamente. En los 3 clasificadores sin over sampling se obtiene p칠simo resultados seg칰n recall; todos valores cercano a 0. Esto ocurre dado que en los 3 casos el clasificador predice de manera muy pobre la clase positiva, donde practicamente todas las predicciones son falsos positivos. 

Este efecto del Precision y Recall ocure claramente por el hecho de que las clases est치n sumamente desbalaceadas, donde la clase positivas corresponde a apenas un 6% de los datos totales, entonces independiente de que se estratificaron el conjunt ode entrenamiento y prueba, estos estar치n sumamente desbalanceados cada uno por si solo. Por esto se puede observar en las matrices de confusi칩n que la clase negativa (jugador no es seleccionado nacional) predice bien, dado que el entrenamiento fue hecho en su mayor칤a con muestras de jugadores de la clase negativa, cuando apenas se entren칩 con un 6% de jugadores de la clase positiva.

El F1-Score combina Precisi칩n y Recall a traves de $\text{F1-Score} = 2 \, \frac{Precision\cdot Recall}{Precision+Recall}$. Claramente para el caso sin over sampling este m칠trica es mala o enga침adora dado los valores de Precision y Recall analizado anteriormente.

Para mejorar los resultados se propone la t칠cnica de Over Sampling, donde se balancean las clases al sobre muestrear de la clase minoritaria hasta alcanzar la misma cantidad de muestras de la clase mayoritaria. Se prob칩 esta t칠cnica con los tres clasificadores y se puede notar la clara mejora de las m칠tricas con valores que tienen m치s sentido para la clase positiva. En particular se hace notar que aunque mejora bastante el resultado, es d칤ficil que pueda mejorar m치s dado que, aunque las clases est칠n balanceadas, el conjunto de entrenamiento contiene muchas muestras de la clase positiva repetidas lo cual no entrega una gran variabilidad en las muestras.

````python
ToNormalStandarScaler = Pipeline([('yeo-johnson', PowerTransformer(method='yeo-johnson')),
                                  ('scaler', StandardScaler())
                                 ])

preprocessing_transformer = ColumnTransformer(
    transformers=[
        ('ReduceCardOneHotEncoder1', OneHotEncoder(sparse=False, min_frequency=0.03, handle_unknown='infrequent_if_exist'),  ['Nationality', 'Club_Position']),
        ('OneHotEncoder1', OneHotEncoder(sparse=False),  ['Work_Rate']),
        ('OneHotEncoder2', OneHotEncoder(sparse=False, drop='if_binary'),  ['Preffered_Foot']),
        ('Identity', IdentityTransformer(),  ['Weak_foot', 'Skill_Moves']),
        ('StandardScaler', StandardScaler(),  ['Height', 'Weight']),
        ('ToNormalStandarScaler', ToNormalStandarScaler, ['Age', 'Ball_Control', 'Dribbling',
                                                          'Marking', 'Sliding_Tackle', 'Standing_Tackle',
                                                          'Aggression', 'Reactions', 'Interceptions',
                                                          'Vision', 'Composure', 'Crossing',
                                                          'Short_Pass', 'Long_Pass', 'Acceleration',
                                                          'Speed', 'Stamina', 'Strength',
                                                          'Balance', 'Agility', 'Jumping',
                                                          'Heading', 'Shot_Power', 'Finishing',
                                                          'Long_Shots', 'Curve', 'Freekick_Accuracy',
                                                          'Penalties', 'Volleys'])
    ])

from sklearn.linear_model import (
    LinearRegression,
    TheilSenRegressor,
    RANSACRegressor,
    HuberRegressor,
)

from sklearn import linear_model
reg = linear_model.Lasso(alpha=0.1)

estimators = [
    ("OLS", LinearRegression()),
    ("Theil-Sen", TheilSenRegressor(random_state=42)),
    ("RANSAC", RANSACRegressor(random_state=42)),
    ("HuberRegressor", HuberRegressor()),
    ("Lasso", linear_model.Lasso(alpha=0.3)),
]

players_pipeline1 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[0]
])

players_pipeline2 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[1]
])

players_pipeline3 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[2]
])

players_pipeline4 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[3]
])

players_pipeline5 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[4]
])
````

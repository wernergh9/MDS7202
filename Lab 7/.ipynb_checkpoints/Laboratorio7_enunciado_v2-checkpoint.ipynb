````python
class IdentityTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, input_array, y=None):
        return self
    
    def transform(self, input_array, y=None):
        return input_array*1
````

````python
ToNormalStandarScaler = Pipeline([('yeo-johnson', PowerTransformer(method='yeo-johnson')),
                                  ('scaler', StandardScaler())
                                 ])

preprocessing_transformer = ColumnTransformer(
    transformers=[
        ('ReduceCardOneHotEncoder1', OneHotEncoder(sparse=False, min_frequency=0.03, handle_unknown='infrequent_if_exist'),  ['Nationality','Club_Position']),
        ('OneHotEncoder1', OneHotEncoder(sparse=False),  ['Work_Rate']),
        ('OneHotEncoder2', OneHotEncoder(sparse=False, drop='if_binary'),  ['Preffered_Foot']),
        ('Identity', IdentityTransformer(),  ['Weak_foot', 'Skill_Moves']),
        ('StandardScaler', StandardScaler(),  ['Height', 'Weight']),
        ('ToNormalStandarScaler', ToNormalStandarScaler, ['Age', 'Ball_Control', 'Dribbling',
                                                          'Marking', 'Sliding_Tackle', 'Standing_Tackle',
                                                          'Aggression', 'Reactions', 'Interceptions',
                                                          'Vision', 'Composure', 'Crossing',
                                                          'Short_Pass', 'Long_Pass', 'Acceleration',
                                                          'Speed', 'Stamina', 'Strength',
                                                          'Balance', 'Agility', 'Jumping',
                                                          'Heading', 'Shot_Power', 'Finishing',
                                                          'Long_Shots', 'Curve', 'Freekick_Accuracy',
                                                          'Penalties', 'Volleys'])
    ])


players_pipeline1 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', svm.SVC())
])

players_pipeline2 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', RandomForestClassifier())
])

players_pipeline3 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', LogisticRegression())
])

players_pipeline1_ros = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', svm.SVC())
])

players_pipeline2_ros = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', RandomForestClassifier())
])

players_pipeline3_ros = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    ('Classifier', LogisticRegression())
])
````

Justificación transformaciones:

* El primer OHE se aplica a variables categóricas con un número amplio de categorías, esto debido a que este OHE crea una categoría donde agrupa todas las categorías poco frecuentes en una. De esta manera se puede reducir en gran medida la dimensionalidad de los datos finales.

* El segundo OHE se aplica a otras variables categóricas no binarias con un número reducido de opciones, a diferencia del anterior donde las opciones eran más de 20 en el caso de Posiciones. En este caso sólo se aplica a la columna work_rate.

* El último OHE se aplica a las variables categóricas binarias, ya que al hacer la transformación a una variable binaria indicando el argumento "if_binary", se obtiene una sola columna con 0s para la primera clase y 1s para la otra (eliminando la necesidad de tener 2 columnas con esto mismo). 

* Luego, a modo de que todas las columnas necesarias estén disponibles, se agrega una transformación identidad. Esto para aquellos datos que no necesitan transformación como Weak_foot y Skill_moves.

* La penúltima transformación es un standar scaler, que se aplica sólo a variables que se ve que distribuyen normal.

* Para aquellas variables que no distribuyen normal, se les aplica una transformación para adaptar las distribuciones seguido del standar scaler.

La precisión indica cuantas predicciones de la clase positiva 1 (jugador es seleccionado nacional) son correctas. Para todos los casos sin over sampling es posible notar que la precisión es muy malo o engañador. Para el caso de SVC su valor es de 0, pero no existe ninguna predicción para la clase positiva. Por otro lado el Random Forest obtiene una precisión de 1, pero es engañador porque observando la matriz de confusión, apenas 2 predicciones fueron para la clase positiva, y calzó que ambas predicciones fueron correctas. Por último ocurre lo mismo para el Logistic Regressor, donde apenas 3 predicciones fueron de la clase positiva.

El Recall indica cuantos de los casos positivos fueron predichos correctamente. En los 3 clasificadores sin over sampling se obtiene pésimo resultados según recall; todos valores cercano a 0. Esto ocurre dado que en los 3 casos el clasificador predice de manera muy pobre la clase positiva, donde practicamente todas las predicciones son falsos positivos. 

Este efecto del Precision y Recall ocure claramente por el hecho de que las clases están sumamente desbalaceadas, donde la clase positivas corresponde a apenas un 6% de los datos totales, entonces independiente de que se estratificaron el conjunt ode entrenamiento y prueba, estos estarán sumamente desbalanceados cada uno por si solo. Por esto se puede observar en las matrices de confusión que la clase negativa (jugador no es seleccionado nacional) predice bien, dado que el entrenamiento fue hecho en su mayoría con muestras de jugadores de la clase negativa, cuando apenas se entrenó con un 6% de jugadores de la clase positiva.

El F1-Score combina Precisión y Recall a traves de $\text{F1-Score} = 2 \, \frac{Precision\cdot Recall}{Precision+Recall}$. Claramente para el caso sin over sampling este métrica es mala o engañadora dado los valores de Precision y Recall analizado anteriormente.

Para mejorar los resultados se propone la técnica de Over Sampling, donde se balancean las clases al sobre muestrear de la clase minoritaria hasta alcanzar la misma cantidad de muestras de la clase mayoritaria. Se probó esta técnica con los tres clasificadores y se puede notar la clara mejora de las métricas con valores que tienen más sentido para la clase positiva. En particular se hace notar que aunque mejora bastante el resultado, es díficil que pueda mejorar más dado que, aunque las clases estén balanceadas, el conjunto de entrenamiento contiene muchas muestras de la clase positiva repetidas lo cual no entrega una gran variabilidad en las muestras.

````python
### Código aquí ###
df_sueldos = pd.read_csv('sueldos.csv')
df_sueldos.drop(['Unnamed: 0'], axis=1, inplace=True)
df_sueldos = df_sueldos.rename(columns = {'Player':'Name'})

new_df = df_players.merge(df_sueldos, on='Name')
````

````python
new_df
````

````python
ToNormalStandarScaler = Pipeline([('yeo-johnson', PowerTransformer(method='yeo-johnson')),
                                  ('scaler', StandardScaler())
                                 ])

preprocessing_transformer = ColumnTransformer(
    transformers=[
        ('ReduceCardOneHotEncoder1', OneHotEncoder(sparse=False, min_frequency=0.03, handle_unknown='infrequent_if_exist'),  ['Nationality', 'Club_Position']),
        ('OneHotEncoder1', OneHotEncoder(sparse=False),  ['Work_Rate']),
        ('OneHotEncoder2', OneHotEncoder(sparse=False, drop='if_binary'),  ['Preffered_Foot']),
        ('Identity', IdentityTransformer(),  ['Weak_foot', 'Skill_Moves']),
        ('StandardScaler', StandardScaler(),  ['Height', 'Weight']),
        ('ToNormalStandarScaler', ToNormalStandarScaler, ['Age', 'Ball_Control', 'Dribbling',
                                                          'Marking', 'Sliding_Tackle', 'Standing_Tackle',
                                                          'Aggression', 'Reactions', 'Interceptions',
                                                          'Vision', 'Composure', 'Crossing',
                                                          'Short_Pass', 'Long_Pass', 'Acceleration',
                                                          'Speed', 'Stamina', 'Strength',
                                                          'Balance', 'Agility', 'Jumping',
                                                          'Heading', 'Shot_Power', 'Finishing',
                                                          'Long_Shots', 'Curve', 'Freekick_Accuracy',
                                                          'Penalties', 'Volleys'])
    ])

from sklearn.linear_model import (
    LinearRegression,
    TheilSenRegressor,
    RANSACRegressor,
    HuberRegressor,
)

from sklearn import linear_model
reg = linear_model.Lasso(alpha=0.1)

estimators = [
    ("OLS", LinearRegression()),
    ("Theil-Sen", TheilSenRegressor(random_state=42)),
    ("RANSAC", RANSACRegressor(random_state=42)),
    ("HuberRegressor", HuberRegressor()),
    ("Lasso", linear_model.Lasso(alpha=0.3)),
]

players_pipeline1 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[0]
])

players_pipeline2 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[1]
])

players_pipeline3 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[2]
])

players_pipeline4 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[3]
])

players_pipeline5 = Pipeline([
    ('Preprocessing', preprocessing_transformer),
    estimators[4]
])
````

````python
### Código aquí ###
X_train, X_test, y_train, y_test = train_test_split(
    new_df.drop(columns=["Weekly Salary"]), new_df['Weekly Salary'], test_size=0.2, shuffle=True)
````

````python
players_pipeline1 = players_pipeline1.fit(X_train, y_train)
players_pipeline2 = players_pipeline2.fit(X_train, y_train)
players_pipeline3 = players_pipeline3.fit(X_train, y_train)
players_pipeline4 = players_pipeline4.fit(X_train, y_train)
players_pipeline5 = players_pipeline5.fit(X_train, y_train)
````

````python
y_pred1 = players_pipeline1.predict(X_test)
y_pred2 = players_pipeline2.predict(X_test)
y_pred3 = players_pipeline3.predict(X_test)
y_pred4 = players_pipeline4.predict(X_test)
y_pred5 = players_pipeline5.predict(X_test)
````

````python
from sklearn.metrics import r2_score


pipes = [players_pipeline1, players_pipeline2, players_pipeline3, players_pipeline4, players_pipeline5]
r2 = []
for pipe in pipes:
    yhat = pipe.predict(X_test)
    r_squared = r2_score(y_test, yhat)
    r2.append(r_squared)

print(r2)
````
